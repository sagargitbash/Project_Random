```python
import numpy as np            #This is all about importing necessary module!
import random 
import matplotlib.pyplot as plt
```

```python

stepsize = 1
numsteps = 100       #(Here we  have initialized all the required variables up to line 4!and position variable
position = [0]       # initially has value 0 and this list is going to increase by appending it with loop generating
x = 0                #variable x.)
for i in range(numsteps):#((here, loop is going on from default value 0 to 100 where 100 is excluded with a step of 1))
    number = np.random.rand()#(((this function generates float vlaue from[0,1) for each value of i with respective loop)))
    if number < 0.5:  #here if condition is applied!
        x = x - stepsize#if above condition is true this will run and assign value to x
    else:
        x = x + stepsize #and if above criteria is not met, this step will assign the value to x
    position.append(x) #Here, above generated value of x for each loop is appended to the position list, initially containing value 0.here appends means to edit the list just by adding the generated value by storing at the end!!!
```

```python
print(position) #This shows the randomly generated position,final list! which has size 101 since 0 was already initialize in the list!!
```

```python
plt.plot(position)#This lineof code runs the plot(graph) of index_0f_position_list as x_value(which is default) and position list as y_value!!!
```

```python
ns = 100                        # no of steps
positions = [0]     # all particles start at x=0
HEAD = 0;  TAIL = 1         # constants

for step in range(ns):
        coin = random.randint(0,1)  # flip coin
        if coin == HEAD:
            positions.append(positions[-1]+1)  # one unit length to the right
        elif coin == TAIL:
            positions.append(positions[-1]-1)   # one unit length to the left
```

```python
plt.plot(positions)
print(positions)
```

```python

N = 1000                   # number of steps
d = 1                      # step length
x = np.zeros(N+1)          # x coordinates 
        
#x[0] = 0        # set initial position

for i in range(0, N):
    r = random.random()         # random number in [0,1)
    if r < 0.5:          
        x[i+1] = x[i] + d
    else:
        x[i+1] = x[i] - d
```
