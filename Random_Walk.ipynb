If we listen the word 'random' ,instantly we think  about the unordered and unpredictable objects and numbers.Nature consists lots of phenomena which is itself spontaneous and random,we cant predict it with full accuracy but somehow there is a way to predict it up to some extent.This approcah/way is called Random Walk.It is basically a simulation process in which movement of some object is determined probabilistically by the help of anlayzing lots of data with predicatability mindset.The randomness in nature is called True Random but in the world of computer though we generate random number that seems  random to we human beings and help to simulate random walk process but such random number is generated based on logic.such types of radomness is called pseudo random.




```python
import numpy as np              
import random           #This section is all about importing necessary modules! Note: Here, though random is imported particularly but it is not necessary if numpy module is already imported!!
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
stepsize = 1         #step size is defined as 1.
numsteps = 100       #(Here we  have initialized all the required variables up to line 4!and position variable
position = [0]       # initially has value 0 and this list size is going to increase by appending it with loop generating
x = 0                #variable x.)
for i in range(numsteps):  #((here, loop is going on from default value 0 to 100 where 100 is excluded with a step of 1))
    number = np.random.rand()  #(((this function generates a float vlaue from[0,1) for each value of i of for loop)))
    if number < 0.5:    #here if condition is applied!
        x = x - stepsize   #if above condition is true this will run and assign value to x
    else:
        x = x + stepsize   #and if above criteria is not met, this step will assign the value to x
    position.append(x)     #Here, above generated value of x for each loop is appended to the position list, initially containing value 0.here appends means to edit the list just by adding the new generated value by storing it at the end !!!
```

```python
print(position)       #This shows the randomly generated position,final list! which has size 101 since 0 was already initialize in the list!!
plt.plot(position)   #This line of code runs the plot(graph) between index position of list as x_parameter(which is default) and position_list value as y_parameter!!!
```

```python
ns = 100                        # no of steps
positions = [0]     # all particles start at x=0
HEAD = 0;  TAIL = 1         # constants

for step in range(ns):
        coin = random.randint(0,1)  # flip coin
        if coin == HEAD:
            positions.append(positions[-1]+1)  # one unit length to the right
        elif coin == TAIL:
            positions.append(positions[-1]-1)   # one unit length to the left
```

```python
plt.plot(positions)
print(positions)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random
HEAD = 1
TAIL = 2




a = np.zeros((2, 1000))
b = a.transpose()
ns = 1000                    # no of steps

for step in range(ns):
        coin_x = random.randint(1,2)

        # Move on x axis
        if coin_x == HEAD:
           b[step,0]+=1  
        elif coin_x == TAIL:
           b[step,0]-=1   

        # Move on y axis
        coin_y = random.randint(1,2)
        if coin_y == HEAD:
           b[step,1]+=1 
        elif coin_y == TAIL:
           b[step,1]-=1 

x = b[:,0].cumsum()
y = b[:, 1].cumsum()
plt.plot(x,y, linestyle = '', marker = '.')
```

```python
import numpy as np
import random
HEAD = 1
TAIL = 2



a = np.zeros((3, 1000))
b = a.transpose()
ns = 1000                       # no of steps

for step in range(ns):
        coin_x = random.randint(1,2)

        # Move on x axis
        if coin_x == HEAD:
           b[step,0]+=1  
        elif coin_x == TAIL:
           b[step,0]-=1   

        # Move on y axis
        coin_y = random.randint(1,2)
        if coin_y == HEAD:
           b[step,1]+=1 
        elif coin_y == TAIL:
           b[step,1]-=1  
        
        #move on z axis
        coin_z = random.randint(1,2)
        if coin_z == HEAD:
           b[step,2]+=1 
        elif coin_z == TAIL:
           b[step,2]-=1 
%matplotlib inline
fig_3d = plt.figure(figsize = (10,8))
fig_3d.add_subplot(1,1,1,projection="3d")
ax = plt.gca()
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.plot(b[:,0].cumsum(),b[:,1].cumsum(),b[:,2].cumsum(), linestyle = '', marker = '.')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
numtrajs = 2000
numsteps = 1000

steps = np.random.random((numtrajs,numsteps))
steps[steps>0.5] = 1
steps[steps<=0.5] = -1
trajs = np.cumsum(steps,axis=1)
plt.figure(figsize = (10,8))
for traj in trajs:
    plt.plot(traj,alpha=0.5)
plt.plot(np.average(trajs,axis=0),color='black',lw=2)
plt.title('individual trajectories')
plt.ylabel(r'$\left<x\right>$')
plt.xlabel('time')
print(steps)
print(trajs)
print(np.average(trajs,axis=0))
plt.show()
```

**3. Day_3: Mean square distance and RootMSD practise portion:** 

```python
import numpy as np
import matplotlib.pyplot as plt
numtrajs = 1000
numsteps = 2000
steps = np.random.random((numtrajs,numsteps))
steps[steps>0.5] = 1
steps[steps<=0.5] = -1
trajs = np.cumsum(steps,axis=1)
new_trajs_squared_Position=np.square(trajs)
x=np.average(new_trajs_squared_Position,axis=0)
ax=plt.figure(figsize = (10,8))
for traj in new_trajs_squared_Position :
    plt.plot(traj,alpha=0.5)
plt.plot(x,color='black',lw=2)
plt.title('MSD')
plt.ylabel(r'$\left<x^2\right>$')
plt.xlabel('time')
print(steps)
print(trajs)
print(new_trajs_squared_Position)
print(x)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
numtrajs = 500
numsteps = 2000
steps = np.random.random((numtrajs,numsteps))
steps[steps>0.5] = 1
steps[steps<=0.5] = -1
trajs = np.cumsum(steps,axis=1)
new_trajs_squared_Position=np.square(trajs)
x=np.average(new_trajs_squared_Position,axis=0)
y=np.sqrt(x)
ax=plt.figure(figsize = (10,8))
for traj in trajs :
    plt.plot(traj,alpha=0.5)
plt.plot(y,color='black',lw=2)
plt.title('RMSD')
plt.ylabel(r'$\left<x\right>$')
plt.xlabel('time')
print(steps)
print(trajs)
print(new_trajs_squared_Position)
print(x)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
numtrajs = 500
numsteps = 2000
times = range(numsteps)
steps = np.random.random((numtrajs,numsteps))
steps[steps>0.5] = 1
steps[steps<=0.5] = -1
trajs = np.cumsum(steps,axis=1)
plt.plot(times,np.mean(trajs**2,axis=0),color='black',lw=2)
plt.title('mean squared displacement', fontsize = 14)
plt.ylabel(r'$\left<x^2\right>$', fontsize = 14)
plt.xlabel('time', fontsize = 14)
```

```python
def msd(x, m, c):
  return m*x+c

parameters, covariance = curve_fit(msd, times, np.mean(trajs**2,axis=0) )
fit_m = parameters[0]
fit_c = parameters[1]
print(fit_m)
print(fit_c)
```

```python
fit_y = msd(times, fit_m, fit_c)
plt.plot(times,np.mean(trajs**2,axis=0) , linestyle ='-',label='data', lw = 2, color = 'black')
plt.plot(times, fit_y, '-', label='fit', color = 'red')
plt.title('means squared displacement', fontsize = 14)
plt.ylabel(r'$\left<x^2\right>$', fontsize = 14)
plt.xlabel('time', fontsize = 14)
plt.legend()
```
