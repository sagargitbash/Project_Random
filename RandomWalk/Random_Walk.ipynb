**1. Day_2: Basis python approaches of plotting 2 and 3 -dimensional random walk plots:**


```python
import matplotlib.pyplot as plt
import numpy as np

numsteps = 2000 #number of steps of particle to go....is initialized
#times = range(numsteps)
steps = np.random.random((numsteps,2)) #this will result into 2000x2 array each element having value between[0,1)
steps[steps>0.5] = 1  # it check  all the  4000  elements in steps array and overwrite its value with 1 if steps>0.5 is satisfied.
steps[steps<=0.5] = -1 # it check  all the  4000  elements in steps array and overwrite its value with -1 if steps<=0.5 is satisfied.
x, y = np.cumsum(steps, axis = 0)[:,0],np.cumsum(steps, axis = 0)[:,1] # this will assign one column elements cumsum into variable x and other colums cumsum into variable y
plt.plot(x,y, linestyle = '', marker = '.') 
plt.xlabel('x_value')
plt.ylabel('y_value')
plt.title('Random Motion in 2D');

print(x)
print(y)
```

```python
#This is the next approach of creating above same 2 D random walk by resoluting its path into x and y axis.
HEAD = 1
TAIL = 2
a = np.zeros((2, 1000)) #This will create 2000 elements 2x1000 array with values 0.0 as float type
b = a.transpose()    # This will just transpose it into 1000x2 array.
ns = 1000             # No of steps is defined
for step in range(ns):  #just a for loop
        coin_x = random.randint(1,2)  # intialize coin_x variable with either 1 or 0 for each loop iteration.
        #--------------------------------------------------------------------------------------#
        if coin_x == HEAD:  #this portion will fill up first column all rows with 1 or -1 by overwriting the respective 0 values
           b[step,0]+=1  
        elif coin_x == TAIL:
           b[step,0]-=1 
        #-------------------------------------------------------------------------------------#  
        coin_y = random.randint(1,2)
        if coin_y == HEAD:
           b[step,1]+=1   #This portion will fill up second column all rows with 1 and -1 by overwriting the respective 0 values
        elif coin_y == TAIL:
           b[step,1]-=1 
        #--------------------------------------------------------------------------------------#
x = b[:,0].cumsum() # it cumsum all rows and 0 colums by index
y = b[:, 1].cumsum() # it cumsum all rows  and 1 colums by index
plt.plot(x,y, linestyle = '', marker = '.') #give commands to plot
```

```python
from mpl_toolkits import mplot3d # this is the required module with submodule mplot3d to plot 3D graphs
numsteps = 2000
#times = range(numsteps)
steps = np.random.random((numsteps,3))
steps[steps>0.5] = 1
steps[steps<=0.5] = -1
x, y, z = np.cumsum(steps, axis = 0)[:,0],np.cumsum(steps, axis = 0)[:,1], np.cumsum(steps, axis = 0)[:,2] # three variable are allocated with cumsum values of their respective columns
fig = plt.figure(figsize = (10,8)) #figure size is allocated in fig variable
ax=fig.add_subplot(1,1,1,projection="3d") #Here add_subplot is inbuilt function to add subplot with rown ,column and index and these all subplot function are added to fig and  all things at the end are stored at variable ax.Here 3d projection means a point is indiacted in space by three resolute axes.
ax = plt.gca() #gca means get current axis  which indicates the current subplot !
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.plot(x,y,z, linestyle = '', marker = '.')
plt.title('Random Motion in 3D');
plt.show()
```

```python
#Another way of 3 dimensional plot of random walk.
HEAD = 1
TAIL = 2
a = np.zeros((3, 1000)) #This will create 3x1000 array each having value 0.0 with float datatype.
b = a.transpose()   #above array changes into  1000x3 array
ns = 1000        #number of steps              
for step in range(ns): #for loop
        coin_x = random.randint(1,2)
        if coin_x == HEAD:
           b[step,0]+=1  
        elif coin_x == TAIL:
           b[step,0]-=1                  #This allocates x-axis,y-axis and z-axis points 
        coin_y = random.randint(1,2)
        if coin_y == HEAD:
           b[step,1]+=1 
        elif coin_y == TAIL:
           b[step,1]-=1  
        coin_z = random.randint(1,2)
        if coin_z == HEAD:
           b[step,2]+=1 
        elif coin_z == TAIL:
           b[step,2]-=1 
%matplotlib inline
fig_3d = plt.figure(figsize = (10,8))  #General  plot functions
fig_3d.add_subplot(1,1,1,projection="3d")
ax = plt.gca()
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.plot(b[:,0].cumsum(),b[:,1].cumsum(),b[:,2].cumsum(), linestyle = '', marker = '.')
plt.show()
```

**3. Day_3: Mean square distance and RootMSD practise portion:** 

In statistical mechanics, the mean squared displacement (MSD)/ average squared displacement/ mean square fluctuation is a measure of the deviation of the position of a particle with respect to a reference position over time. It can be thought of as measuring the portion of the system "explored" by the random walker. In the realm of biophysics and environmental engineering, the Mean Squared Displacement is measured over time to determine if a particle is spreading solely due to diffusion, or if an advective(convection) force is also contributing.
$$MSD(t)= <(x_t -x_0)^2>$$ i.e. averae of square of difference between initil and final position of random particles relative to time!

```python
import numpy as np              
import random           #This section is all about importing necessary modules! Note: Here, though random is imported particularly but it is not necessary if numpy module is already imported!!
import matplotlib.pyplot as plt
from google.colab import drive  #Linking onedrive with google colab,so that different path and location  of drive can be accessible !
drive.mount('/content/drive/') 
import seaborn as sns #Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.Here,it
sns.set_theme()       #used to set default theme of seaborn library
numtrajs = 1000  #Number of trajectories is defined and also can be called as number of particles defined!
numsteps = 2000  #number of steps is defined here
steps = np.random.random((numtrajs,numsteps)) # 1000X2000 array is created having value between [0,1) as float datatype
steps[steps>0.5] = 1  # it check  all the  2000000  elements in steps array and overwrite its value with 1 if steps>0.5 is satisfied.
steps[steps<=0.5] = -1 # it check  all the  2000000  elements in steps array and overwrite its value with -1 if steps<=0.5 is satisfied.
trajs = np.cumsum(steps,axis=1)  # #this cumsum along all columns in each row
new_trajs_squared_Position=np.square(trajs) #this square the  array trajs containing cumsum
x=np.average(new_trajs_squared_Position,axis=0) #it find average along all rows of each colum through out the array
ax=plt.figure(figsize = (10,10)) #figure size is defined 
for traj in new_trajs_squared_Position : #loop starts here
    plt.plot(traj,alpha=0.5)             #it plot trajectories for each loop with transparency half,since it is inside the loop
plt.plot(x,color='black',lw=2) #it plot MSD 
plt.title('MSD')
plt.ylabel(r'$\left<x^2\right>$')
plt.xlabel('times(x)')
print(steps)
print(trajs)
print(new_trajs_squared_Position)
print(x)
plt.show()
```

**The above MSD trajectories show that if number of particles increased increasing number of trajectories ,the MSD plot shows that the particles seems to go far form the initial position slowly with increasing time.**

```python
#Now we are going to line_fit the above MSD data plot
from scipy.optimize import curve_fit #from optimize submodule of scipy module  curve_fit is imported to fit the data
def msd(x, m, c): #a function named msd is defined with parameters x,m,c
  return m*x+c   # and msd=m*x+c

parameters, covariance = curve_fit(msd, times, np.mean(trajs**2,axis=0) ) # This is a syntax that give us slope and intercepts inside parameters and covariance  of data inside covariance. 
fit_m = parameters[0] #since slope value is located at first index of parameters,it shows the slope of fitted curve.
fit_c = parameters[1]  #since intercept value is located at second index of parameters,it shows the y-intercept of fitted curve.
print(fit_m)
print(fit_c)
```

```python
fit_y = msd(times, fit_m, fit_c) #object named fit_y is created
plt.plot(times,np.mean(trajs**2,axis=0) , linestyle ='-',label='data', lw = 2, color = 'black') #it creates MSD data plot
plt.plot(times, fit_y, '-', label='fit', color = 'red') #it creates the best fit line to our MSD data plot
plt.title('means squared displacement', fontsize = 14)
plt.ylabel(r'$\left<x^2\right>$', fontsize = 14)
plt.xlabel('times',fontsize = 14)
plt.legend()
```

```python
#Parameters for the Brome Mosaic Virus:Brome mosaic virus (BMV) is a positive-strand RNA virus that infects cereal plants, causing mosaic symptoms and stunting(prevent from growing or developing properly).
import numpy as np              
import random           #This section is all about importing necessary modules!
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
m = 7.6e-21      # kg
gamma = 3.4e10   # s^-1
dt = 10e-9       # s
T = 300          # K

kB = 1.380649e-23 # J/K          #Here we have initialize the required parameters of the Brome

max_steps = 5000
x0 = 0           #starting position is defined

sigma = np.sqrt(2*kB*T * dt / (m * gamma))  # m 
sigma
```

To account for the right size for the fluctuations, the random force must be uncorrelated Gaussian noise with zero mean and variance.The variance is given by:
$$\sigma^2 = \frac{2 k_B T \Delta t}{m \gamma}$$
